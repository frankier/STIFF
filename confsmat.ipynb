```python
import pickle
from scikitplot.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
```

```python
confsmat = pickle.load(open('confsmat4.pkl', 'rb'))
confsmat['act'] = list(map(lambda x: {'simp': 'Simple', 'trad': 'Traditional', 'neit': 'Neither'}[x], confsmat['act']))
```

```python
ax = plot_confusion_matrix(confsmat['pred'], confsmat['act'], true_labels=['zh_CN', 'zh_TW'], pred_labels=['Simple', 'Traditional', 'Neither'])

plt.xlabel('OpenCC based classification')
plt.ylabel('OpenSubtitles2018 language code');
plt.savefig('confmat.pdf')
```
